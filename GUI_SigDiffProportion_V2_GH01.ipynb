```python

"""
==============================================================================================
            Creating a GUI  to determine if the difference between proportions is significant
==============================================================================================
GUI_SigDiffPorpotion_v1
            - OBJECTIVE create a GUI   to determine if  the difference in porportions of two
                populations  is significant
                - UPDATES - update the  code used for the sample size program to determine
                if significant differences  exist between proportions

                -cCreate enter button for the fourth parameter

GUI_SigDiffPorpotion_v2
            - OBJECTIVE -  make the code operational
            - UPDATES - ll references to SampleSioze  replaced with SigDiff
            - unnecessary - method declration, i.e. def(self) removed for now
            - RESULTS - SUCCESS! -  three scenarios are  calculated successfully
                - CAVEAT - the  SUBMIT  button does not appear on  the laptop screen,  it does
                appear on the monitor screen. Apparently the laptop screen is too small.
                Needs further investigation.


========================================================================
"""

import tkinter as tk
import math
from tkinter import *
from tkinter import messagebox
from tkinter import ttk

"""
========================================================================
Set up GUI and get inputs
========================================================================
"""

root = tk.Tk()
root.title("Significant Differences of Proportions")
root.geometry("300x200")
# root.geometry("300x1200")
# root.geometry("1300x200")

padx_constant = 20
pady_constant = 20

# Create an overall frame
frame = tk.Frame(root)
frame.pack()


# Creaate  the first user input frame - User_info_frame_1

class SigDiff_Class:

    def __init__(self, root):

        # ------------------------------------------------ Initialize variables -----------------------------------------

        self.Frame_font_tuple = ("Arial", 12, "bold")         # Reference: See https://www.geeksforgeeks.org/how-to-set-font-for-text-in-tkinter/

        # ----------------------------------------------- Dataframe 0 -----------------------------------------------------
        # Dataframe 3 - SAmple Size Message

        self.user_info_frame_S0 = tk.LabelFrame(frame,  text="Significant Difference of Proportions Overview",  font=self.Frame_font_tuple)
        self.user_info_frame_S0.grid(row=0, column=0, sticky="news", padx=padx_constant, pady=pady_constant)                                                # 'Sticky' - see Hala video around 30:00

        self.SigDiffMsg_0 = "Use to answer: Does a common treatment in two different populations yield significantly different results?"                           # it required about six tabs  for this line to align with the one below
        self.msg_label3_0 = tk.Label(self.user_info_frame_S0, text = self.SigDiffMsg_0)
        self.msg_label3_0.grid(row=0, column=0, sticky=tk.W)                                                                                                # Justify to the West, i.e. Left Justify

        self.SigDiffMsg_1 = "Use to ansesr: Do two differnt treatments in the same population yield different responses?"
        self.msg_label3_1 = tk.Label(self.user_info_frame_S0, text = self.SigDiffMsg_1)
        self.msg_label3_1.grid(row=1, column=0, sticky=tk.W)


        # ----------------------------------------------- Initialize Scenario Variables -----------------------------------------------------
        # Initialize input variable labels
        self.First_Param    = "Control: Response"
        self.Second_Param   = "Cotrol: Sample"
        self.Third_Param    = "Offer: Response"
        self.Fourth_Param    = "Offer: Population"

        self.Result_Msg = "Sample SizeX: "

        print(f"\nIn main..")

        # Create and initialize input data variables

        # Scenario 1
        self.in_dat1_1 = IntVar()                           # first input data field is float
        self.in_dat2_1 = IntVar()                           # second input data field is integer
        self.in_dat3_1 = IntVar()                           # third input data field is integer
        self.in_dat4_1 = IntVar()                           # third input data field is integer

        self.in_dat1_1.set(100)
        self.in_dat2_1.set(200)
        self.in_dat3_1.set(500)
        self.in_dat4_1.set(1000)

        # Scenario 2
        self.in_dat1_2 = IntVar()                           # first input data field is float
        self.in_dat2_2 = IntVar()                           # second input data field is integer
        self.in_dat3_2 = IntVar()                           # third input data field is integer
        self.in_dat4_2 = IntVar()                           # third input data field is integer

        self.in_dat1_2.set(100)
        self.in_dat2_2.set(200)
        self.in_dat3_2.set(500)
        self.in_dat4_2.set(1000)


        # Scenario 3
        self.in_dat1_3 = IntVar()                           # first input data field is float
        self.in_dat2_3 = IntVar()                           # second input data field is integer
        self.in_dat3_3 = IntVar()                           # third input data field is integer
        self.in_dat4_3 = IntVar()                           # third input data field is integer

        self.in_dat1_3.set(100)
        self.in_dat2_3.set(200)
        self.in_dat3_3.set(500)
        self.in_dat4_3.set(1000)



        self.Calc_Result = DoubleVar()                      # to hold the result of the calculations
        self.Res_S1 = DoubleVar()                           # to hold the result of the calculations
        self.Res_S2 = DoubleVar()                           # to hold the result of the calculations
        self.Res_S3 = DoubleVar()                           # to hold the result of the calculations


        print(f"\n Line before SUBMIT button.")
        self.submit_button = tk.Button(root, text="SUBMIT", command=self.run_submit)
        self.submit_button.pack(pady=20)                 # double check what 'pady' does'

        # -------------------------------------------------- Overall Run Structure ---------------------------------------------------------

    def run_SigDiff(self):
        print(f"\nIn run method.")

        self.mycounter = 0

        # SCenario 1
        self.user_info_frame = tk.LabelFrame(frame,  text="Significant Difference of Proportions: Scenario 1", font=self.Frame_font_tuple)
        self.user_info_frame.grid(row=1, column=0, padx=padx_constant, pady=pady_constant)
        self.SigDiffPrep(self.in_dat1_1, self.in_dat2_1, self.in_dat3_1, self.in_dat4_1)
        self.SigDiffCalc(self.in_dat1_1, self.in_dat2_1, self.in_dat3_1, self.in_dat4_1)
        # print(f"\n ~~~ In run_SigDiff, Scenario {self.mycounter}:  MargOfError_ is {self.MargOfError_},  TargetProp_  is {self.TargetProp_}, Confidence level is {self.Conf_level}.")
        # print(f"\n ~~~ In run_SigDiff:  Sample size result for Scenario {self.mycounter} is {self.Calc_Result}.")

        # SCenario 2
        self.user_info_frame = tk.LabelFrame(frame,  text="Significant Difference of Proportions: Scenario 2", font=self.Frame_font_tuple)
        self.user_info_frame.grid(row=2, column=0, padx=padx_constant, pady=pady_constant)
        self.SigDiffPrep(self.in_dat1_2, self.in_dat2_2, self.in_dat3_2, self.in_dat4_2)
        self.SigDiffCalc(self.in_dat1_2, self.in_dat2_2, self.in_dat3_2, self.in_dat4_2)
        # print(f"\n ~~~ In run_SigDiff, Scenario {self.mycounter}:  MargOfError_ is {self.MargOfError_},  TargetProp_  is {self.TargetProp_}, Confidence level is {self.Conf_level}.")
        # print(f"\n ~~~ In run_SigDiff:  Sample size result for Scenario {self.mycounter} is {self.Calc_Result}.")

        # SCenario 3
        self.user_info_frame = tk.LabelFrame(frame,  text="Significant Difference of Proportions: Scenario 3", font=self.Frame_font_tuple)
        self.user_info_frame.grid(row=3, column=0, padx=padx_constant, pady=pady_constant)
        self.SigDiffPrep(self.in_dat1_3, self.in_dat2_3, self.in_dat3_3, self.in_dat4_3)
        self.SigDiffCalc(self.in_dat1_3, self.in_dat2_3, self.in_dat3_3, self.in_dat4_3)
        # print(f"\n ~~~ In run_SigDiff, Scenario {self.mycounter}:  MargOfError_ is {self.MargOfError_},  TargetProp_  is {self.TargetProp_}, Confidence level is {self.Conf_level}.")
        # print(f"\n ~~~ In run_SigDiff:  Sample size result for Scenario {self.mycounter} is {self.Calc_Result}.")


    # Prepare entry and message widgets for input variables
    def SigDiffPrep(self, in_dat1, in_dat2, in_dat3, in_dat4):
        print (f"\n\nP-P-P In SigDiffPrep and my counter is {self.mycounter}.")

        self.msg_label = tk.Label(self.user_info_frame, text = self.First_Param)
        self.msg_label.grid(row=0, column=0)

        self.msg_label = tk.Label(self.user_info_frame, text = self.Second_Param)
        self.msg_label.grid(row=0, column=1)

        self.msg_label = tk.Label(self.user_info_frame, text = self.Third_Param)
        self.msg_label.grid(row=2, column=0)

        self.msg_label = tk.Label(self.user_info_frame, text = self.Fourth_Param)
        self.msg_label.grid(row=2, column=1)


        self.Enter1 = tk.Entry(self.user_info_frame, width=20, textvariable=in_dat1)
        self.Enter1.grid(row=1, column=0)

        self.Enter2 = tk.Entry(self.user_info_frame, width=20, textvariable=in_dat2)
        self.Enter2.grid(row=1, column=1)

        self.Enter3 = tk.Entry(self.user_info_frame, width=20, textvariable=in_dat3)
        self.Enter3.grid(row=3, column=0)

        self.Enter4 = tk.Entry(self.user_info_frame,  width=20, textvariable=in_dat4)
        self.Enter4.grid(row=3, column=1)

        # create an output to store the results
        self.result_message_1 = tk.Message(self.user_info_frame, width=200, text=self.Result_Msg, fg="blue")
        self.result_message_1.grid(row=3, column=2)

        print(f"\n In SigDiffPrep: In_dat1 is {self.in_dat1_1} and In_dat2 is {self.in_dat2_1}.")

        # space out all fields using user_info_frame
        for widget in self.user_info_frame.winfo_children():
            widget.grid_configure(padx=10, pady=5)

    # --------------------------------------------  Prop[ortion Size Method -----------------------------------

    # Calculate proportions, numerator, denominator and  z_score proportion

    def SigDiffCalc(self, in_dat1, in_dat2, in_dat3, in_dat4):

        # User supplied data - Four input variables


        """
        self.Target_1   = 1070
        self.Pop_1      = 10000

        self.Target_2   = 2000
        self.Pop_2      = 20000
        """

        self.Target_1   = in_dat1.get()
        self.Pop_1      = in_dat2.get()

        self.Target_2   = in_dat3.get()
        self.Pop_2      = in_dat4.get()


        # initialize variables

        self.SigLevel_99 = 2.576
        self.SigLevel_95 = 1.960
        self.SigLevel_90 = 1.645

        # Calculate z-scores

        self.Prop_1 = self.Target_1/ self.Pop_1
        self.Prop_2 = self.Target_2/ self.Pop_2
        self.PropHat = (self.Target_1 + self.Target_2)/(self.Pop_1 + self.Pop_2)

        self.Numerator = self.Prop_1 - self.Prop_2
        self.Denom =  math.sqrt(self.PropHat * (1-self.PropHat) * ((1/self.Pop_1) + (1/self.Pop_2)))
        self.Z_Score_Prop = abs(self.Numerator/ self.Denom)
        self.OutputResults()


    def OutputResults(self):
        print(f"\n All clear!")
        print(f"\n The Z score is {self.Z_Score_Prop}, Proportion 1 is {self.Prop_1}, Proportion 2 is {self.Prop_2}, and the denominator is {self.Denom}")

        self.Res_message_99 = "Groups are significant at the 99 pct level of confidence."
        self.Res_message_95 = "Groups are significant at the 95 pct level of confidence."
        self.Res_message_90 = "Groups are significant at the 90 pct level of confidence."
        self.Res_message_sub_90 = "Groups are not significant at the 90 pct level of confidence."

        if self.Z_Score_Prop >= self.SigLevel_99:
            self.result_message_1.config(text=f"{self.Res_message_99}")
            print(f"\n Groups are significantly different at the 99% level of confidence.")

        elif self.Z_Score_Prop >= self.SigLevel_95:
            self.result_message_1.config(text=f"{self.Res_message_95}")
            print(f"\n Groups are significantly different at the 95% level of confidence.")

        elif self.Z_Score_Prop >= self.SigLevel_90:
            self.result_message_1.config(text=f"{self.Res_message_90}")
            print(f"\n Groups are significantly different at the 90% level of confidence.")

        else:
            self.result_message_1.config(text=f"{self.Res_message_sub_90}")
            print(f"\n Groups are not significantly different at the 90% level of confidence.")

        print(f"\n DIAGNOSTICS: The Z score is {self.Z_Score_Prop}, Proportion 1 is {self.Prop_1}, Proportion 2 is {self.Prop_2}, and the denominator is {self.Denom}")


    # --------------------------------------------- Sample Siae Calc -------------------------------------------

    """
    def SampleSizeCalc(self, in_dat1, in_dat2, in_dat3):
        print(f"\n\nC-C-C In SampleSizeCalc method and My counter is {self.mycounter}.")


        self.MargOfError = in_dat1.get()                # 2.10
        self.Conf_level  = in_dat2.get()                # 90, 95, 95
        self.TargetProp  = in_dat3.get()                # 50

        self.MargOfError_ = self.MargOfError/ 100
        self.TargetProp_ = self.TargetProp/ 100
        self.n_SampleSize_ = 0
        self.ConfLevelFlag = False


        self.Conf_level_Dct     = {80:1.28, 85:1.44, 90:1.645, 95:1.960, 99:2.576, 99.9:3.291}                    # Confidence level dictionary

        print(f"\nIn SampleSizeCalc method, after variable declaration, before calculation.")
        print(f"\nIn SAmpleSizeCalc: In_dat1 (Margin of error) is {self.in_dat1_1} and In_dat2 (Confidence level )is {self.in_dat2_1}.")
        print(f"\nMargin of error is {self.MargOfError} and Cofnince level is {self.Conf_level}.")

        for key, value in self.Conf_level_Dct.items():                    # Get the z-Score equivalent to the confidence level inputted
            if  self.Conf_level ==  key:
                self.z_Score = value
                self.ConfLevelFlag = True
                print(f"\nCofnidence level is {self.Conf_level} and z_Score is  {self.z_Score}.")
            else:
                # do nothing, leave the Confidence level flag false
                pass


        if self.ConfLevelFlag == True:
            self.Calc_Result = round((self.z_Score**2 * (self.TargetProp_ * (1 - self.TargetProp_))) / (self.MargOfError_)**2)
            self.result_message_1.config(text=f"Sample Size: {self.Calc_Result}")
            print(f"\n +++ In scenario {self.mycounter} smargin of error is {self.MargOfError_}.")

    """


    def run_submit(self):
        print (f"\n\n>>> Executing run_submit method.")
        self.run_SigDiff()
        print(f"\n --- My counter is {self.mycounter}, Calc_Result is {self.Calc_Result}.")


if __name__ == "__main__":
    ExecSigDiff = SigDiff_Class(root)
    ExecSigDiff.run_SigDiff()
    root.mainloop()
```
